```python
import pandas as pd
df = pd.read_csv('/content/Post-COVID_Conditions.csv')
display(df.describe())
```

```python
import random

N = len(df)
n = 100
k = N // n

start_index = random.randint(0, k - 1)

sampled_df = df.iloc[start_index::k]

display(sampled_df.head())
```

```python
unique_states = df['State'].unique()

num_clusters_to_sample = 10
sampled_clusters = random.sample(list(unique_states), num_clusters_to_sample)

cluster_sampled_df = df[df['State'].isin(sampled_clusters)]

display(cluster_sampled_df.head())
print(f"\nOriginal DataFrame shape: {df.shape}")
print(f"Cluster Sampled DataFrame shape: {cluster_sampled_df.shape}")
```
